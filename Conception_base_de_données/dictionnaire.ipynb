# Dictionnaires des données

| Table           | Nom             | Type          | Description |
|-----------------|-----------------|---------------|-------------|
| UTILISATEUR     | id              | INT           | Identifiant unique de l’utilisateur (PK) |
| UTILISATEUR     | email           | VARCHAR(180)  | Email de l’utilisateur |
| UTILISATEUR     | lastname        | VARCHAR(120)  | Le nom de l'utilisateur |
| UTILISATEUR     | firstname       | VARCHAR(120)  | Le prénom de l’utilisateur |
| UTILISATEUR     | phone           | VARCHAR(30)   | Le numéro de téléphone de l'utilisateur |
| UTILISATEUR     | password        | VARCHAR(255)  | Mot de passe de l’utilisateur |
| UTILISATEUR     | roles           | JSON          | Rôles de l’utilisateur |
| HOTEL           | id              | INT           | Identifiant unique de l’hôtel (PK) |
| HOTEL           | name            | VARCHAR(255)  | Nom de l’hôtel |
| HOTEL           | city            | VARCHAR(255)  | Ville où se situe l’hôtel |
| HOTEL           | address         | VARCHAR(255)  | Adresse complète de l’hôtel |
| CHAMBRE         | id              | INT           | Identifiant unique de la chambre (PK) |
| CHAMBRE         | number          | INT           | Numéro de la chambre |
| CHAMBRE         | pricePerNight   | FLOAT         | Prix par nuit de la chambre |
| CHAMBRE         | id_hotel        | INT           | FK vers HOTEL(id), indique à quel hôtel la chambre appartient |
| SERVICE         | id              | INT           | Identifiant unique du service (PK) |
| SERVICE         | name            | VARCHAR(255)  | Nom du service |
| SERVICE         | description     | VARCHAR(255)  | Description du service |
| SERVICE         | price           | FLOAT         | Prix du service |
| RESERVATION     | id              | INT           | Identifiant unique de la réservation (PK) |
| RESERVATION     | startDate       | DATETIME      | Date de début de la réservation |
| RESERVATION     | endDate         | DATETIME      | Date de fin de la réservation |
| RESERVATION     | status          | VARCHAR(255)  | Statut de la réservation |
| RESERVATION     | id_utilisateur  | INT           | FK vers UTILISATEUR(id), indique quel utilisateur a réservé |
| RESERVATION     | id_chambre      | INT           | FK vers CHAMBRE(id), indique quelle chambre est réservée |
| MAINTENANCE     | id              | INT           | Identifiant unique de la maintenance (PK) |
| MAINTENANCE     | startDate       | DATETIME      | Date de début prévue de la maintenance |
| MAINTENANCE     | endDatePlanned  | DATETIME      | Date de fin prévue de la maintenance |
| MAINTENANCE     | endDateReal     | DATETIME      | Date de fin réelle de la maintenance (peut être NULL) |
| MAINTENANCE     | status          | VARCHAR(255)  | Statut de la maintenance |
| MAINTENANCE     | description     | TEXT          | Description des travaux à effectuer |
| MAINTENANCE     | id_chambre      | INT           | FK vers CHAMBRE(id), indique sur quelle chambre la maintenance est effectuée |
| CHAMBRE_SERVICE | id_chambre      | INT           | FK vers CHAMBRE(id), identifie la chambre |
| CHAMBRE_SERVICE | id_service      | INT           | FK vers SERVICE(id), identifie le service |
| CHAMBRE_SERVICE | (PK composée)   | —             | La combinaison id_chambre + id_service fait la PK |
